```python
travel = df[df['gb3']=='여행']
```

```python
trv_rate = travel['v2'].value_counts().to_frame()         # 여행 목적지 순위 
```

```python
trv_rate = trv_rate.rename(columns = {'v2':'freq'})
```

```python
trv_rate['v2'] = trv_rate.index
```

```python
trv_rate.plot.bar()
```

```python
Trpt = df[df['gb2']=='교통']
```

```python
Trpt
```

```python
Trpt['cost'] = Trpt['vlm']/Trpt['usec']
```

```python
Trpt['cost']
```

```python
mean_trans = Trpt.groupby('v2')['cost'].agg(**{'cost':'mean'}).reset_index() #
# 여행지별 교통비 평균 
```

```python
mean_trans
```

```python
mean_trans = mean_trans.sort_values(by='cost' ,ascending=False)
```
